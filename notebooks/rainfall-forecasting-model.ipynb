# Rainfall Prediction Classifier | Rain in Australia

<b>Install and import the required libraries<b/>

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import seaborn as sns
```

<b>Load the data<b/>

```python
#  Datasets available at Kaggle https://www.kaggle.com/datasets/jsphyg/weather-dataset-rattle-package/
df= pd.read_csv("../data/weatherAUS_2.csv")
df.head()
```

```python
df.count()
```

<b>Drop all rows with missing values<b/>

```python
# Drop all rows with missing values to simplify preprocessing
df = df.dropna()

# Check the remaining number of observations and data types
df.info()
```

```python
# Display all column names to review available features
df.columns
```

```python
# Rename columns to reflect the new target of predicting today's rain using past data
df = df.rename(columns={'RainToday': 'RainYesterday',
                        'RainTomorrow': 'RainToday'
                        })
```

```python
# Filter the dataset to include only data from three nearby locations: Melbourne, Melbourne Airport, and Watsonia
# These locations are within ~20 km of each other and are expected to share similar weather patterns

df = df[df.Location.isin(['Melbourne','MelbourneAirport','Watsonia'])]
```

```python
# Display updated dataset info to confirm we still have enough observations to build a localized model
df.info()
```

<b>Create a function to map dates to seasons:<b/>

```python
# Define a function to extract the season from a given date
# Australia’s seasons:
# Summer: Dec, Jan, Feb | Autumn: Mar, Apr, May | Winter: Jun, Jul, Aug | Spring: Sep, Oct, Nov

def date_to_season(date):
    month = date.month
    if (month == 12) or (month == 1) or (month == 2):
        return 'Summer'
    elif (month == 3) or (month == 4) or (month == 5):
        return 'Autumn'
    elif (month == 6) or (month == 7) or (month == 8):
        return 'Winter'
    elif (month == 9) or (month == 10) or (month == 11):
        return 'Spring'
```

```python
# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Apply the function to the 'Date' column to create a new 'Season' column
df['Season'] = df['Date'].apply(date_to_season)

# Drop the original 'Date' column as it's no longer needed
df = df.drop(columns='Date')

df
```

<b>Feature and target dataframe:<b/>

```python
# Define the features by dropping the target column
X = df.drop(columns='RainToday', axis=1)

# Define the target variable
y = df['RainToday']
```

```python
# Display the number of occurrences for each class in the target variable
y.value_counts()
```

-Rain is less frequent: It rains on fewer days than it doesn't in Melbourne.

-Imbalanced dataset: The "No Rain" class dominates.

-Naive accuracy is misleading: Always predicting "No" gives high accuracy but poor usefulness.

-Next steps: Handle imbalance using resampling, class weights, or better metrics like F1-score.

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
# Use stratify=y to maintain class distribution (rain/no rain) in both sets
# test_size=0.2 means 20% of data is for testing
# random_state=42 ensures results are reproducible
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
# Select numerical columns by their data types (e.g., int, float)
numeric_features = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Select categorical columns by their data types (e.g., object, category)
categorical_features = X_train.select_dtypes(include=['object', 'category']).columns.tolist()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

# Transformer for numeric features: scale them using StandardScaler
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

# Transformer for categorical features: one-hot encode with ignore for unknown categories
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Combine both transformers into a single preprocessing step
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)
```

```python
# Combine numeric and categorical transformers into one preprocessing step
preprocessor = ColumnTransformer(
    transformers=[
        # Apply numeric_transformer to numeric features
        ('num', numeric_transformer, numeric_features),
        # Apply categorical_transformer to categorical features
        ('cat', categorical_transformer, categorical_features)
    ]
)
```

# Pipeline: Preprocessing + Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

pipeline = Pipeline(steps=[
    # Apply preprocessing steps defined earlier
    ('preprocessor', preprocessor),
    # Add Random Forest classifier as the final estimator
    ('classifier', RandomForestClassifier(random_state=42))
])
```

```python
# Define parameter grid for GridSearchCV

param_grid = {
    'classifier__n_estimators': [50, 100],        # Number of trees
    'classifier__max_depth': [None, 10, 20],      # Max depth of trees
    'classifier__min_samples_split': [2, 5]       # Min samples required to split a node
}
```

```python
# - Define stratified cross-validation

from sklearn.model_selection import StratifiedKFold

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)  # Stratified 5-fold CV with shuffling
```

# Instantiate and fit GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

# Create GridSearchCV with pipeline, params, stratified CV, accuracy metric, verbose output
grid_search = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    cv=cv,
    scoring='accuracy',
    verbose=2
)

# Fit GridSearchCV on the training data to find best hyperparameters
grid_search.fit(X_train, y_train)
```

# Print best parameters and cross-validation score

```python
# Display the best parameters found by grid search
print("\nBest parameters found: ", grid_search.best_params_)

# Display the best mean accuracy score from cross-validation
print("Best cross-validation score: {:.2f}".format(grid_search.best_score_))
```

# Model's estimated score

```python
# Evaluate the best model on the test set and print the accuracy score
test_score = grid_search.score(X_test, y_test)
print("Test set score: {:.2f}".format(test_score))
```

# Get the model predictions from the grid search estimator on the unseen data

```python
# Get predictions for the test set using the best model found by grid search
y_pred = grid_search.predict(X_test)
```

# Print the classification report

```python
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

# Plot the confusion matrix

```python
# Compute confusion matrix from true and predicted labels
conf_matrix = confusion_matrix(y_test, y_pred)

# Create a visualization object for the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix)

# Plot the confusion matrix with a blue color map
disp.plot(cmap='Blues')

# Add a title to the plot
plt.title('Confusion Matrix')

# Display the plot
plt.show()
```

<b>Extract Feature Importances<b/>

```python
# Extract the feature importances from the best estimator (Random Forest) inside the pipeline
feature_importances = grid_search.best_estimator_['classifier'].feature_importances_
```

<b>Plot Top N Feature Importances<b/>

```python
# Combine numeric and one-hot encoded categorical feature names in the same order as model input
feature_names = numeric_features + list(
    grid_search.best_estimator_['preprocessor']
    .named_transformers_['cat']
    .named_steps['onehot']
    .get_feature_names_out(categorical_features)
)

# Create a DataFrame of feature importances
importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': feature_importances
}).sort_values(by='Importance', ascending=False)

# Set how many top features to display
N = 20
top_features = importance_df.head(N)

# Plot the top N most important features
plt.figure(figsize=(10, 6))
plt.barh(top_features['Feature'], top_features['Importance'], color='skyblue')
plt.gca().invert_yaxis()  # Most important feature appears on top
plt.title(f'Top {N} Most Important Features in Predicting Rain Today')
plt.xlabel('Importance Score')
plt.show()
```

# Update pipeline, estimator, and parameter grid

<b>Logistic Regression Model<b/>

```python
# Replace the classifier in the pipeline with Logistic Regression
pipeline.set_params(classifier=LogisticRegression(random_state=42))

# Update the estimator in the GridSearchCV to use the new pipeline
grid_search.estimator = pipeline

# Define a new parameter grid specific to Logistic Regression
param_grid = {
    'classifier__solver': ['liblinear'],
    'classifier__penalty': ['l1', 'l2'],
    'classifier__class_weight': [None, 'balanced']
}

# Assign the new parameter grid to GridSearchCV
grid_search.param_grid = param_grid
```

<b>Fit the model and make predictions<b/>

```python
# Fit the updated pipeline with Logistic Regression to the training data
model = grid_search.fit(X_train, y_train)

# Predict on the test set using the best found model
y_pred = model.predict(X_test)
```

# Comparing Model Performances

```python
from sklearn.metrics import classification_report, confusion_matrix

# Print the classification report for Logistic Regression
print("Classification Report - Logistic Regression:\n")
print(classification_report(y_test, y_pred))

# Generate and plot the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, cmap='Blues', fmt='d')
plt.title('Confusion Matrix - Logistic Regression')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.tight_layout()
plt.show()
```
